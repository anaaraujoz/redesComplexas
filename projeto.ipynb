# Explorando as redes

```python
# Setup das bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import networkx as nx
import itertools
```

```python
# Read the edgelist file
baywetG = nx.read_weighted_edgelist("eco-foodweb-baywet.txt")
baydryG = nx.read_weighted_edgelist("eco-foodweb-baydry.txt")

pos = nx.spring_layout(baywetG)
colors = range(20)
options = {
    "node_color": "#A0CBE2",
    "edge_color": colors,
    "width": 4,
    "edge_cmap": plt.cm.Blues,
    "with_labels": True,
}
nx.draw(baywetG, pos, **options)
plt.show()
#nx.draw(baywetG, pos,with_labels=True, node_color='skyblue', node_size=250, edge_color='gray', font_size=8)
```
