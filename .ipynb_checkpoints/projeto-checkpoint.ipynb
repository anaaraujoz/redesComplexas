# Explorando as redes

```python
# Setup das bibliotecas
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import networkx as nx
import itertools
```

```python
# Read the edgelist file
baywetG = nx.read_weighted_edgelist("eco-foodweb-baywet.txt")
baydryG = nx.read_weighted_edgelist("eco-foodweb-baydry.txt")
```

```python
plt.figure(figsize=(16,16))
pos = nx.spring_layout(baywetG, seed=63)

#node_sizes = [3 + 10 * i for i in range(len(baywetG))]
M = baywetG.number_of_edges()
print
edge_colors = range(2, M + 2)
edge_alphas = [(5 + i) / (M + 4) for i in range(M)]
cmap = plt.cm.Blues

nodes = nx.draw_networkx_nodes(baywetG, pos, node_size=20, node_color="k")
labels = nx.draw_networkx_labels(baywetG, pos, verticalalignment="bottom")
edges = nx.draw_networkx_edges(
    baywetG,
    pos,
    node_size=10,
    arrowstyle="->",
    arrowsize=1,
    edge_color=edge_colors,
    edge_cmap=cmap,
    width=2,
    arrows=True
)
# set alpha value for each edge
for i in range(M):
    edges[i].set_alpha(edge_alphas[i])

pc = mpl.collections.PatchCollection(edges, cmap=cmap)
pc.set_array(edge_colors)

ax = plt.gca()
ax.set_axis_off()
#plt.colorbar(pc, ax=ax)
plt.show()
```
